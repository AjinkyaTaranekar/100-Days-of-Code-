# Data Operators in R
# There are mainly 5 different types of operators, which are listed below:

# Arithmetic Operators: 
# Perform arithmetic operations such as:
addition, subtraction, multiplication, division etc.
# Assignment Operators: 
Assignment operators are used to assign values. For example:
Assignment Operator =
# Syntax:
variable name = value
<ul></li>
<li>>x=5</li>
<li>>x</li>
<li>Output : [1] 5</li>
    
# Assignment Operator <-
# Syntax:
variable name <- value
<ul>
<li> x<<- 15</li>
<li>>x</li>
<li>Output : [1] 15</li>
    
# Assignment Operator<<-
# Syntax :
variable name <<- value
<li>>x <<- 2</li>
<li>>x</li>
<li>Output : [1] 2</li>

# Assignment Operator ->
# Syntax :
value -> variable name
<li>>25 -> x</li>
<li>>x</li>
<li>Output : [1] 25</li>

# 3. Relational Operator:
It defines a relation between two entities. For example: <,>,<=,!= etc.

<li>> x < -3 </li>
<li>>x!=2</li>
<li>Output: [1] TRUE</li>

# 4. Logical Operators: 
These operators compare the two entities and are typically used with boolean (logical) values such as &, | and !. 

<li>> x<-2 </li>
<li> >2&3 </li>
<li>Output: [1] TRUE</li>

# 5. Special Operators:
These operators are used for specific purpose, not for logical computation. For example:

# It creates the series of numbers in sequence for a vector.

<li> >x<-2:8 </li>
<li>>x</li>
<li>Output: [1] 2 3 4 5 6 7 8</li>

# %in% This operator is used to identify if an element belongs to a vector.
Example:
<li> >x <- 2:8 </li>
<li>>y<-5</li>
<li>>y %in% x</li>
<li>Output : [1] TRUE</li>
